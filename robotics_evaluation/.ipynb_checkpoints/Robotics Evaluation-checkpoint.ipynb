```python
def load_data(files):
    data = np.array((1,7))
    for f in files:
        loaded = np.loadtxt(f)
        try:
            data = np.vstack((loaded,data))
        except:
            data = loaded
    return data
```

```python
def round_to_multiple(number, multiple):
    return multiple * np.round(number / multiple)

round_to_multiple(np.random.rand(10), 0.25)
```

```python
from time import time

def create_reference(folder, round_multiple):
    experiments = glob.glob(f"{folder}/*")
    uniques = np.array(1)
    all_data = np.array(1)
    for e in experiments:
        files = glob.glob(f"{e}/*local*samples*")
        data = load_data(files)
        if len(data) == 2:
            continue
        rounded_current = round_to_multiple(data, round_multiple)
        unis = np.unique(rounded_current,axis=0)
        try:
            uniques = np.unique(np.vstack((uniques,unis)),axis=0)
        except:
            uniques = unis
        try:
            all_data = np.vstack((all_data,data))
            rounded = np.vstack((rounded,rounded_current))
        except:
            rounded = rounded_current
            all_data = data
    i = 0
    ref = np.zeros(uniques.shape)
    start = time()
    print("uniques = ", len(uniques))
    for u in uniques:
        if i%1000 == 0:
            print(i)
            print(time() - start)
            start = time()
        not_rounded_select = all_data[np.all(np.equal(rounded,u),axis=1)]
        l = len(not_rounded_select)
        idx = np.random.randint(l)
        ref[i,:] = not_rounded_select[idx]
        i += 1
    return ref

#ref3 = create_reference("3dof",0.0025)        
ref7 = create_reference("7dof",0.5)

#np.savetxt("3dof_ref.dat",ref3,fmt='%f')
np.savetxt("7dof_ref.dat",ref7,fmt='%f')

```

```python
#coverage calculation
import numba
from  numba import jit,njit

@njit(nogil=True)
def dist(ref, samples):
    n,m = ref.shape
    l = len(samples)
    lowest = np.Inf
    current = 0
    min_vals = np.zeros(n)
    for i in range(n):
        row = ref[i]
        for k in range(l):
            for j in range(m):
                current += (row[j] - samples[k,j])**2
            if current < lowest:
                lowest = current
            current = 0
        min_vals[i] = lowest
        lowest = np.Inf
    return min_vals


def adts(ref, samples):
    vals = dist(ref,samples)
    return np.mean(vals)
```

```python
print("------------------------")
folder = "7dof"
ref = np.loadtxt(f"{folder}_ref.dat")
experiments = glob.glob(f"{folder}/*")
coverages = {}
for e in experiments:
    files = glob.glob(f"{e}/*local*samples*")
    if len(files) == 0:
        continue
    samples = load_data(files)
    try:
        out = dist(ref,samples)
        print(get_name(exp_map,e), " = ",np.mean(out))
    except:
        out = dist(ref,samples)
        print(e, " = ",np.mean(out))
    coverages[e] = out

        
        
```

```python
for k,v in coverages.items():
    np.savetxt(f"{k}_coverages.dat",v,fmt='%f')
```

```python
# num iterations


folder = ["7dof","3dof"]
for f in folder:
    experiments = glob.glob(f"{f}/*")
    for e in experiments:
        if "probs" in e:
            continue
        files = glob.glob(f"{e}/*num*")
        if len(files) > 0:
            iterations = 0
            for f in files:
                iterations += np.loadtxt(f).sum()
            files = glob.glob(f"{e}/*local*samples*")
            n_samples = 0
            for f in files:
                n_samples += np.loadtxt(f).shape[0]
        print("ADTS for ", e, " = ", iterations/n_samples)
        
    
```
